```python
import random
class DesChiffresEtDesLettres:

    def __init__(self): #This method defines small numbers and large numbers
        self.small_numbers = list(range(1, 11))
        self.large_numbers = [25, 50, 75, 100]

    def generate_target_number(self): #This method generates a random 3-digit target number
        return random.randint(100, 999)

    def generate_six_numbers(self): #This method selects 4 small numbers and 2 large numbers randomly
        return random.sample(self.small_numbers, 4) + random.sample(self.large_numbers, 2)

    def evaluate_expression(self, expression, available_numbers): #This method evaluates the player's expression by ensuring all numbers used are from the available numbers and checking the validity of the arithmetic expression.
        try:
            #Extract digits from the expression and check if they match the available numbers
            numbers_used = [int(x) for x in expression.split() if x.isdigit()]
            #If the numbers match the available ones, evaluate the expression
            if sorted(numbers_used) == sorted(available_numbers):
                return eval(expression) #Evaluate using Python's eval()
        except:
            return None #Return None if there's any error in evaluation

    def calculate_number_round_score(self, target, guess): #This method calculates the score based on how close the guess is to the target
        return max(100 - abs(target - guess) // 10, 0)

    def generate_letter(self, choice): #This method defines vowels and consonants
        vowels, consonants = "AEIOU", "BCDFGHJKLMNPQRSTVWXYZ"
        #Return a random vowel or consonant based on the player's choice
        return random.choice(vowels if choice == 'vowel' else consonants)

    def prompt_for_letters(self): #This method prompts the player to select vowels or consonants 10 times, and returns the list of generated letters.
        return [self.generate_letter(input("Choose 'vowel' or 'consonant': ")) for _ in range(10)]

    def get_longest_word(self, letters, word_list): #This method finds the longest word that can be formed using the given letters. Instead of using .count(), manually track the letters used
        def can_form_word(word, available_letters):
            #Copy the available letters so we can modify the list while checking the word
            temp_letters = available_letters.copy()
            #Check if each letter in the word can be formed using the available letters
            for letter in word:
                if letter in temp_letters:
                    temp_letters.remove(letter) #Remove letter from available letters if found
                else:
                    return False #If letter can't be found, the word can't be formed
            return True
        #Find the longest word that can be formed
        longest_word = ""
        for word in word_list:
            #Check if the word can be formed and if it's longer than the current longest
            if can_form_word(word.upper(), letters) and len(word) > len(longest_word):
                longest_word = word #Update the longest word
        return longest_word

    def calculate_word_round_score(self, longest_word_length, player_word_length): #This method calculates score based on the length of the player's word vs the longest word
        return max(100 - (longest_word_length - player_word_length) * 10, 0)

    def load_words_from_file(self, filename): #This method loads words from the specified file and returns them as a list. If the file is not found, return an empty list.
        try:
            with open(filename, 'r') as file:
                return [line.strip() for line in file]
        except FileNotFoundError:
            print(f"Error: The file {filename} was not found.")
            return []
game = DesChiffresEtDesLettres()
#NUMBERS ROUND
target = game.generate_target_number()
numbers = game.generate_six_numbers()
print(f"Target: {target}\nNumbers: {numbers}")
expression = input("Enter an arithmetic expression using the numbers: ")
guess = game.evaluate_expression(expression, numbers)
if guess:
    print(f"Your guess: {guess}, Score: {game.calculate_number_round_score(target, guess)}")

#LETTERS ROUND
letters = game.prompt_for_letters()
print(f"Your letters: {''.join(letters)}")
word_list = game.load_words_from_file('words.txt')
if word_list:
    longest_word = game.get_longest_word(letters, word_list)
    player_word = input("Enter your word: ").upper()
    print(f"Longest word: {longest_word}\nYour word: {player_word}, Score: {game.calculate_word_round_score(len(longest_word), len(player_word))}")
```
